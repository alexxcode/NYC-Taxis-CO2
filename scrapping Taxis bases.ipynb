```python
import requests
from bs4 import BeautifulSoup
import urllib.parse
import pandas as pd
import os

# URL de la pgina web de la TLC de Nueva York
url = "https://www.nyc.gov/site/tlc/about/tlc-trip-record-data.page"

# Realiza una solicitud GET a la pgina web
response = requests.get(url)

# Verifica que la solicitud sea exitosa
if response.status_code == 200:
    # Analiza el contenido HTML de la pgina con BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")

    # Busca enlaces a archivos PARQUET en la pgina web
    links = soup.find_all("a", href=True)
    
    # Filtra los enlaces que apuntan a archivos PARQUET
    parquet_links = [link["href"] for link in links if link["href"].endswith(".parquet")]

    # Inicializa una lista para guardar las URLs
    urls = []

    # Recorre los enlaces encontrados y agrega las URLs a la lista
    for parquet_link in parquet_links:
        # Construye la URL completa si el enlace es relativo
        full_url = urllib.parse.urljoin(url, parquet_link)
        urls.append(full_url)

    # Imprime las URLs encontradas
    for url in urls:
        print(f"Enlace PARQUET encontrado: {url}")

else:
    print("La solicitud no fue exitosa. Verifica la URL o la conexin a internet.")
```

```python
# Directorio de salida para los archivos Parquet
output_directory = 'Bases_parquet'

# Crear el directorio de salida si no existe
os.makedirs(output_directory, exist_ok=True)

# Descargar y guardar los archivos Parquet
for url in urls:
    file_name = os.path.join(output_directory, os.path.basename(url))
    
    # Verificar si el archivo ya existe en el directorio
    if not os.path.exists(file_name):
        try:
            urllib.request.urlretrieve(url, file_name)
            print("Archivo Parquet descargado con xito:", file_name)
        except urllib.error.HTTPError as e:
            print(f"Error al descargar {url}: {e}")
    else:
        print("El archivo Parquet ya existe en el directorio:", file_name)

print("Proceso de descarga completado.")
```
