```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

# A√±adir la columna 'pandemia' al DataFrame
yellow_data['pandemia'] = ((yellow_data['year'] == 2020) & (yellow_data['month'] >= 3) | 
                           (yellow_data['year'] > 2020) & (yellow_data['year'] < 2023) | 
                           (yellow_data['year'] == 2023) & (yellow_data['month'] < 5)).astype(int)

# Definir la funci√≥n para ajustar el modelo SARIMA, incluyendo la pandemia como variable ex√≥gena
def fit_sarima_model_with_exog(data, target_column, exog_column, order=(2, 1, 2), seasonal_order=(2, 1, 2, 12), forecast_periods=12):
    sarima_results = {}
    
    for borough in data['pickup_borough'].unique():
        borough_ts = data[(data['pickup_borough'] == borough)].copy()
        borough_ts['date'] = pd.to_datetime(borough_ts['year'].astype(str) + '-' + borough_ts['month'].astype(str))
        borough_ts.set_index('date', inplace=True)
        
        # Definir la serie temporal y la variable ex√≥gena
        ts = borough_ts[target_column]
        exog = borough_ts[exog_column]
        
        # Ajustar el modelo SARIMA con variable ex√≥gena
        sarima_model = SARIMAX(ts, exog=exog, order=order, seasonal_order=seasonal_order, 
                                enforce_stationarity=False, enforce_invertibility=False)
        sarima_result = sarima_model.fit(disp=False)
        
        # Crear las fechas de pron√≥stico futuras
        last_date = ts.index[-1]
        future_dates = pd.date_range(start=last_date + pd.offsets.MonthBegin(1), periods=forecast_periods, freq='MS')
        
        # Crear la serie de la variable ex√≥gena para el pron√≥stico (0 despu√©s de mayo de 2023)
        exog_future = pd.Series(0, index=future_dates)
        exog_future.loc[:pd.Timestamp('2023-05')] = 1  # La pandemia termina en mayo de 2023
        
        # Realizar el pron√≥stico
        forecast = sarima_result.get_forecast(steps=forecast_periods, exog=exog_future)
        forecast_series = forecast.predicted_mean
        forecast_conf_int = forecast.conf_int()
        
        # Almacenar resultados
        sarima_results[borough] = {
            'model': sarima_result,
            'forecast': forecast_series,
            'conf_int': forecast_conf_int
        }
        
        # Graficar resultados
        plt.figure(figsize=(10, 5))
        plt.plot(ts, label='Observed')
        plt.plot(future_dates, forecast_series, label='Forecast')
        plt.fill_between(future_dates, forecast_conf_int.iloc[:, 0], forecast_conf_int.iloc[:, 1], color='pink', alpha=0.3)
        plt.axvline(x=pd.Timestamp('2020-03'), color='orange', linestyle='--', label='Start of Pandemic')
        plt.axvline(x=pd.Timestamp('2023-05'), color='green', linestyle='--', label='End of Pandemic')
        plt.title(f'{borough} - CO2 Emissions Forecast with Pandemic Effect-YellowTaxi')
        plt.xlabel('Date')
        plt.ylabel(target_column)
        plt.legend()
        plt.show()
    
    return sarima_results

# Ajustar el modelo SARIMA para las emisiones de CO2 de cada pickup_borough
co2_results = fit_sarima_model_with_exog(yellow_data, 'co2_month', 'pandemia')
```

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

# Cargar datos
yellow_data_path = 'Final_yellow_co_ruido_costo.csv' 
yellow_data = pd.read_csv(yellow_data_path)

# A√±adir la columna 'pandemia' al DataFrame
yellow_data['pandemia'] = ((yellow_data['year'] == 2020) & (yellow_data['month'] >= 3) |
                           (yellow_data['year'] > 2020) & (yellow_data['year'] < 2023) |
                           (yellow_data['year'] == 2023) & (yellow_data['month'] < 5)).astype(int)

# Definir la funci√≥n para ajustar el modelo SARIMA y hacer predicciones
def fit_sarima_and_forecast_co2(data, exog_column, order, seasonal_order, forecast_periods):
    sarima_results = {}
    for borough in data['pickup_borough'].unique():
        borough_data = data[data['pickup_borough'] == borough].copy()
        borough_data['date'] = pd.to_datetime(borough_data['year'].astype(str) + '-' + borough_data['month'].astype(str))
        borough_data.set_index('date', inplace=True)
        ts = borough_data['co2_month']
        exog = borough_data[exog_column]
        
        # Ajustar el modelo SARIMA
        model = SARIMAX(ts, exog=exog, order=order, seasonal_order=seasonal_order, enforce_stationarity=False, enforce_invertibility=False)
        results = model.fit(disp=False)
        
        # Pronosticar el futuro
        future_dates = pd.date_range(start=ts.index[-1] + pd.offsets.MonthBegin(1), periods=forecast_periods, freq='MS')
        exog_future = pd.Series(0, index=future_dates)
        exog_future.loc[:pd.Timestamp('2023-05')] = 1
        forecast = results.get_forecast(steps=forecast_periods, exog=exog_future)
        
        # Almacenar resultados
        sarima_results[borough] = forecast.predicted_mean
    
    return sarima_results

# Ajustar el modelo SARIMA y hacer predicciones para los pr√≥ximos 12 meses
forecast_periods = 12
sarima_forecasts_co2 = fit_sarima_and_forecast_co2(yellow_data, 'pandemia', (2, 1, 1), (1, 1, 2, 12), forecast_periods)

# Aplicar la reducci√≥n a las predicciones
electric_vehicle_percent = 20
reduction_factor = electric_vehicle_percent / 100
for borough, forecast_series in sarima_forecasts_co2.items():
    fleet_reduction = (forecast_series) * reduction_factor
    additional_reduction = co2_reduction_coef
    adjusted_forecast = forecast_series - abs(fleet_reduction) - abs(additional_reduction)
    
    # Graficar las predicciones y las l√≠neas ajustadas para CO2
    plt.figure(figsize=(14, 7))
    plt.plot(forecast_series.index, forecast_series, label='Forecast')
    plt.plot(adjusted_forecast.index, adjusted_forecast, label='Forecast Adjusted for Electric Vehicle Conversion', linestyle='--')
    plt.title(f'{borough} - CO2 Forecast Adjusted for Electric Vehicle Conversion')
    plt.xlabel('Date')
    plt.ylabel('CO2 Emissions')
    plt.legend()
    plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt
final_yellow_data = pd.read_csv('Final_yellow_co_ruido_costo.csv')
regresion_ruido = pd.read_csv('regresion_edmunds_ruido.csv')

# A√±adir la columna 'pandemia' al DataFrame
final_yellow_data['pandemia'] = ((final_yellow_data['year'] == 2020) & (final_yellow_data['month'] >= 3) |
                                 (final_yellow_data['year'] > 2020) & (final_yellow_data['year'] < 2023) |
                                 (final_yellow_data['year'] == 2023) & (final_yellow_data['month'] < 5)).astype(int)

# Definir la funci√≥n para ajustar el modelo SARIMA y hacer predicciones para el ruido
def fit_sarima_and_forecast_noise(data, exog_column, order, seasonal_order, forecast_periods):
    sarima_results = {}
    for borough in data['pickup_borough'].unique():
        borough_data = data[data['pickup_borough'] == borough].copy()
        borough_data['date'] = pd.to_datetime(borough_data['year'].astype(str) + '-' + borough_data['month'].astype(str))
        borough_data.set_index('date', inplace=True)
        ts = borough_data['noise_month']  # Usando 'noise_month' para el ruido
        exog = borough_data[exog_column]
        
        # Ajustar el modelo SARIMA
        model = SARIMAX(ts, exog=exog, order=order, seasonal_order=seasonal_order, enforce_stationarity=False, enforce_invertibility=False)
        results = model.fit(disp=False)
        
        # Pronosticar el futuro
        future_dates = pd.date_range(start=ts.index[-1] + pd.offsets.MonthBegin(1), periods=forecast_periods, freq='MS')
        exog_future = pd.Series(0, index=future_dates)
        exog_future.loc[:pd.Timestamp('2023-05')] = 1  # La pandemia termina en mayo de 2023
        forecast = results.get_forecast(steps=forecast_periods, exog=exog_future)
        
        # Almacenar resultados
        sarima_results[borough] = forecast.predicted_mean
    
    return sarima_results

# Ajustar el modelo SARIMA y hacer predicciones para el ruido para los pr√≥ximos 12 meses
forecast_periods = 12
sarima_forecasts_noise = fit_sarima_and_forecast_noise(final_yellow_data, 'pandemia', (2, 1, 2), (2, 2, 1, 12), forecast_periods)

# Cargar el coeficiente de reducci√≥n de ruido de los modelos de regresi√≥n
noise_reduction_coef = -1 * regresion_ruido[regresion_ruido['Unnamed: 0'] == 'Es_Electrico']['Coef'].iloc[0]

# Aplicar la reducci√≥n a las predicciones de ruido
electric_vehicle_percent = 20
reduction_factor_noise = electric_vehicle_percent / 100
for borough, forecast_series in sarima_forecasts_noise.items():
    fleet_reduction_noise = forecast_series * reduction_factor_noise
    adjusted_forecast_noise = forecast_series - (abs(fleet_reduction_noise * noise_reduction_coef))
    
    # Graficar las predicciones y las l√≠neas ajustadas para el ruido
    plt.figure(figsize=(14, 7))
    plt.plot(forecast_series.index, forecast_series, label='Forecast')
    plt.plot(adjusted_forecast_noise.index, adjusted_forecast_noise, label='Forecast Adjusted for Electric Vehicle Conversion', linestyle='--')
    plt.title(f'{borough} - Noise Forecast Adjusted for Electric Vehicle Conversion')
    plt.xlabel('Date')
    plt.ylabel('Noise Level')
    plt.legend()
    plt.show()

# Continuar√© con los siguientes pasos despu√©s de revisar estos resultados.
sarima_forecasts_noise.keys() 
```

```python
import pandas as pd

def export_sarima_noise_results_to_csv(sarima_forecasts_noise, final_data, forecast_periods, electric_vehicle_percent, noise_reduction_coef):
    # Inicializar la lista para almacenar los datos de todas las predicciones de ruido
    all_boroughs_noise_data = []

    # Obtener la √∫ltima fecha de los datos para comenzar las predicciones desde ese punto
    last_date = final_data[['year', 'month']].drop_duplicates().nlargest(1, ['year', 'month'])
    last_date = pd.to_datetime(f"{last_date.iloc[0]['year']}-{last_date.iloc[0]['month']}-01")

    # Calcular las fechas de predicci√≥n
    prediction_dates_noise = pd.date_range(start=last_date, periods=forecast_periods, freq='MS')

    # Iterar sobre las predicciones de ruido de cada borough y almacenar los datos
    for borough, forecast in sarima_forecasts_noise.items():
        forecast_df = pd.DataFrame({
            'predicted_noise': forecast
        })
        forecast_df['date'] = prediction_dates_noise
        forecast_df['year'] = forecast_df['date'].dt.year
        forecast_df['month'] = forecast_df['date'].dt.month
        forecast_df['borough'] = borough

        # Calcular la reducci√≥n del ruido
        reduction_factor_noise = electric_vehicle_percent / 100
        adjusted_forecast_noise = forecast_df['predicted_noise'] - (abs(forecast_df['predicted_noise'] * reduction_factor_noise * noise_reduction_coef))
        
        forecast_df['predicted_noise_adjusted'] = adjusted_forecast_noise
        forecast_df['noise_reduction_coef'] = noise_reduction_coef
        forecast_df['total_noise_reduction'] = abs(forecast_df['predicted_noise'] - adjusted_forecast_noise)

        # Agregar los datos al listado general
        all_boroughs_noise_data.append(forecast_df)

    # Combinar todos los datos en un solo DataFrame
    all_noise_predictions_df = pd.concat(all_boroughs_noise_data)

    # Exportar a CSV
    csv_noise_output_path = 'predicted_noise_yellow_paraPowerBi.csv'
    all_noise_predictions_df.to_csv(csv_noise_output_path, index=False)

    return csv_noise_output_path

# Llamar a la funci√≥n de exportaci√≥n
csv_noise_output_path = export_sarima_noise_results_to_csv(sarima_forecasts_noise, final_yellow_data, forecast_periods, electric_vehicle_percent, noise_reduction_coef)
print(csv_noise_output_path)
```

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

# Parte 1: C√°lculo del MPG Promedio
# Cargar los datos de veh√≠culos
processed_edmunds_data = pd.read_csv('processed_edmunds_data.csv')

# Calcular el MPG promedio para veh√≠culos de gasolina (no el√©ctricos)
gasoline_vehicles = processed_edmunds_data[processed_edmunds_data['Es_Electrico'] == 0]
avg_mpg_gasoline = gasoline_vehicles['City MPG'].mean()

# Calcular el MPG promedio para veh√≠culos el√©ctricos
electric_vehicles = processed_edmunds_data[processed_edmunds_data['Es_Electrico'] == 1]
avg_mpg_electric = electric_vehicles['City MPG'].mean()

# Parte 2: Ajuste del Modelo SARIMA y Visualizaci√≥n del Ahorro
# Cargar los datos
final_yellow_data = pd.read_csv('Final_yellow_co_ruido_costo.csv')

# Definir la variable 'pandemia'
final_yellow_data['pandemia'] = ((final_yellow_data['year'] == 2020) & (final_yellow_data['month'] >= 3) |
                                 (final_yellow_data['year'] > 2020) & (final_yellow_data['year'] < 2023) |
                                 (final_yellow_data['year'] == 2023) & (final_yellow_data['month'] < 5)).astype(int)

# Funci√≥n para ajustar el modelo SARIMA y visualizar el ahorro
def fit_sarima_model_and_plot(data, target_column, exog_column, order=(2, 1, 1), seasonal_order=(1, 1, 2, 12), forecast_periods=12, conversion_percentage=0.20, efficiency_index=1/3):
    sarima_results = {}
    
    for borough in data['pickup_borough'].unique():
        borough_ts = data[(data['pickup_borough'] == borough)].copy()
        borough_ts['date'] = pd.to_datetime(borough_ts['year'].astype(str) + '-' + borough_ts['month'].astype(str))
        borough_ts.set_index('date', inplace=True)
        
        # Serie temporal y variable ex√≥gena
        ts = borough_ts[target_column]
        exog = borough_ts[exog_column]
        
        # Ajustar el modelo SARIMA
        sarima_model = SARIMAX(ts, exog=exog, order=order, seasonal_order=seasonal_order, 
                                enforce_stationarity=False, enforce_invertibility=False)
        sarima_result = sarima_model.fit(disp=False)
        
        # Fechas de pron√≥stico futuras
        last_date = ts.index[-1]
        future_dates = pd.date_range(start=last_date + pd.offsets.MonthBegin(1), periods=forecast_periods, freq='MS')
        
        # Serie de la variable ex√≥gena para el pron√≥stico
        exog_future = pd.Series(0, index=future_dates)
        exog_future.loc[:pd.Timestamp('2023-05')] = 1
        
        # Realizar el pron√≥stico
        forecast = sarima_result.get_forecast(steps=forecast_periods, exog=exog_future)
        forecast_series = forecast.predicted_mean
        
        # Calcular el costo con ahorro por conversi√≥n a veh√≠culos el√©ctricos
        forecast_series_ahorro = forecast_series - (abs(forecast_series * conversion_percentage) / efficiency_index)
        
        # Almacenar resultados
        sarima_results[borough] = {
            'model': sarima_result,
            'forecast': forecast_series,
            'forecast_with_savings': forecast_series_ahorro
        }
        
        # Graficar resultados con ahorro
        plt.figure(figsize=(10, 5))
        plt.plot(forecast_series, label='Pron√≥stico de Costo (Convencionales)')
        plt.plot(forecast_series_ahorro, label='Pron√≥stico con Ahorro (20% El√©ctricos)', linestyle='--')
        plt.title(f'{borough} - Comparaci√≥n de Costos')
        plt.xlabel('Fecha')
        plt.ylabel('Costo')
        plt.legend()
        plt.show()
    
    return sarima_results

# Ajustar el modelo SARIMA para cada borough y visualizar el ahorro
results_with_savings = fit_sarima_model_and_plot(final_yellow_data, 'cost_month', 'pandemia')

# Mostrar los resultados del MPG promedio
avg_mpg_gasoline, avg_mpg_electric
```

```python
# Funci√≥n para exportar los resultados del modelo SARIMA a CSV
def export_sarima_results_to_csv(sarima_results, forecast_periods, conversion_percentage=0.20, efficiency_index=1/3):
    all_boroughs_cost_data = []  # Lista para almacenar los datos de todas las predicciones de costo

    # Obtener la √∫ltima fecha de los datos para comenzar las predicciones desde ese punto
    last_date = final_yellow_data[['year', 'month']].drop_duplicates().nlargest(1, ['year', 'month'])
    last_date = pd.to_datetime(f"{last_date.iloc[0]['year']}-{last_date.iloc[0]['month']}-01")

    # Calcular las fechas de predicci√≥n
    prediction_dates = pd.date_range(start=last_date, periods=forecast_periods, freq='MS')

    # Iterar sobre las predicciones de costo de cada borough y almacenar los datos
    for borough, result in sarima_results.items():
        forecast = result['forecast']
        forecast_with_savings = forecast - (abs(forecast * conversion_percentage) / efficiency_index)

        forecast_df = pd.DataFrame({
            'predicted_cost': forecast,
            'predicted_cost_with_savings': forecast_with_savings
        })
        forecast_df['date'] = prediction_dates
        forecast_df['year'] = forecast_df['date'].dt.year
        forecast_df['month'] = forecast_df['date'].dt.month
        forecast_df['borough'] = borough

        # Calcular la diferencia entre las columnas 'predicted_cost' y 'predicted_cost_with_savings'
        forecast_df['Total_reduction'] = forecast_df['predicted_cost'] - forecast_df['predicted_cost_with_savings']

        # Agregar los datos al listado general
        all_boroughs_cost_data.append(forecast_df)

    # Combinar todos los datos en un solo DataFrame
    all_cost_predictions_df = pd.concat(all_boroughs_cost_data)

    # Exportar a CSV con un nombre espec√≠fico
    csv_output_path = 'predict_ahorro_yellow_PowerBi.csv'
    all_cost_predictions_df.to_csv(csv_output_path, index=False)

    return csv_output_path

# Exportar los resultados del modelo SARIMA a un archivo CSV
csv_file_path = export_sarima_results_to_csv(results_with_savings, 12)
print(csv_file_path)
```
